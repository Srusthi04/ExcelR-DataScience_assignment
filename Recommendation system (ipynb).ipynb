```python
def give_reco(customer_id):
    tem = list(user_sim_df.sort_values([customer_id],ascending=False).head(100).index)
    movie_list=[]
    for i in tem:
        movie_list=movie_list+list(data[data['user_id']==i]['title'])
    return set(movie_list)-set(data[data['user_id']==customer_id]['title'])
```
